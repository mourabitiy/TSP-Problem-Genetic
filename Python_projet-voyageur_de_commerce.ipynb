```python
import time
from itertools import permutations
import numpy as np
import matplotlib.pyplot as plt
import random
from sys import maxsize
from matplotlib.pyplot import figure

def permuter(array):
    if len(array) == 1:
        return [array]
    permutations = []
    for i in range(len(array)): 
        # get all perm's of subarray w/o current item
        perms = permuter(array[:i] + array[i+1:])  
        for p in perms:
            permutations.append([array[i], *p]) 
    return permutations

def distance(x, y, ordre):
    if len(ordre) < len(x):
        raise Exception("Permutation non valide")
    total = 0
    for current_point_index, next_point_index in zip(ordre, ordre[1:]):
        x0, y0 = x[current_point_index], y[current_point_index]
        x1, y1 = x[next_point_index], y[next_point_index]
        total += ((x0 - x1) ** 2 + (y0 - y1) ** 2) ** 0.5
    return total

def trace(x,y,duration,distanceMin):
    from matplotlib.pyplot import figure
    plt.plot(x,y,"o",label= "le chemin le plus court ")
    for i in range(len(x)):
        plt.text(x[0]+0.5, y[0]+0.5,"Ville de départ ", fontsize=8)
        plt.text(x[i], y[i],i, bbox=dict(boxstyle="round",facecolor='red', alpha=0.5), fontsize=12)
    plt.plot(x + [ x[0] ], y + [ y[0] ], "g--")
    plt.scatter(x[0], y[0], c="green", marker=">")
    plt.xlabel("coord. x")
    plt.ylabel("coord. y")
    plt.title("Chemin plus court(Duration : "+str(duration)+" seconds)"+"\n distance : "+str(np.round(distanceMin,2)) + "Km", fontsize=15)
    fig = plt.gcf()
    fig.set_size_inches(10, 10)
    plt.show()

def resolution_exacte(nb):
    debut = time.time()
    #Coordonnées aléatoires des villes 
    x = np.random.randint(20, size=nb)
    y = np.random.randint(20, size=nb)
    villes =[]
    for i in range(0,nb):
        villes.append(i)
    print("=======================================================================")
    print("Coordonnées Générés : ",end="\n")
    for i in range(nb):
        print("Ville ",i+1,": (",x[i], ", ",y[i] ,")")
    cpt = 1
    distanceMin = maxsize
    print("Voulez-vous afficher toutes les permutations possible ?(O/N) : ")
    choix = input()
    if choix =="O" or choix == "o":
        print("Affichage : ")
        for perm in permuter(villes):
                        perm.append(perm[0]) #Ajoute de la ville de départ à la fin du chemin
                        print("Chemin N° " ,cpt,": ",perm)
                        cpt+=1
                        print("Longueur du chemin : ", distance(x,y,perm))
                        if distanceMin>distance(x,y,perm) :
                            distanceMin=distance(x,y,perm)
                            chemin=perm
    else : 
        print("=============== Chemins restants à parcourir : ================ ")
        for perm in permuter(villes):
            perm.append(perm[0])
            cpt+=1
            print("\033[1m"+"`==========> " +str(np.math.factorial(nb)-cpt)+"permutations restantes ",end="\r")
            if distanceMin>distance(x,y,perm):
                distanceMin = distance(x,y,perm)
                chemin = perm
    print(str(nb) + " villes ===> "+str(np.math.factorial((nb))) + "  " + "permutations générée")
    print("=======================================================================================")
    print("\t\t\t============distance Optimal==============  ")
    print(np.round(distanceMin,2) ,"Km")
    print("\t\t\t============Chemin Optimal ===============  ")
    for i in range(nb):
        if i == nb-1:
            print(chemin[i]," : ",x[chemin[i]],",", y[chemin[i]], end = "\n")
        else:
            print(chemin[i]," : ", x[chemin[i]],", ", y[chemin[i]],"-->", end ="\n")
    fin = time.time()
    
    duration = fin - debut
    #listes des coordonnées dans l'ordre le plus court trouvé
    x_optimal=[0]*nb
    y_optimal = [0]*nb
    for i in range(nb):
        x_optimal[chemin[i]]=x[chemin[i]]
        y_optimal[chemin[i]]=y[chemin[i]]
    trace(x_optimal, y_optimal, duration,distanceMin)
    
#####################################################____Main()____###################################################
nb = int(input("Entrez le nombre des villes à génerer aléatoirement : "))
duration = resolution_exacte(nb)
```

Un algorithme génétique est constitué de 5 grandes phases :
    <li>

    <li> Evaluation de la "fitness" (qualité) des individus</li>
<li>Sélection des parents</li>
<li>Création de nouveaux individus : crossover + mutation</li>
<li>Ajout des nouveaux individus dans la population</li>
<li>Retour à la seconde étape</li>

```python
import numpy as np
import random
import math
from sys import maxsize
import matplotlib.pyplot as plt
TAUXMUTATION =10

nbVilles = int(input("Entrez le nombre des villes à générer : "))
Villes = np.random.rand(nbVilles, 2) * 5
nbVilles = len(Villes)

class Population():
    chromosomes = []
    fitness = maxsize

def creer_population(nbVilles):
    individus = []
    for i in range(nbVilles):
        tmp = Population()
        tmp.chromosomes = random.sample(range(1, nbVilles), nbVilles - 1)
        tmp.chromosomes.append(0)
        tmp.fitness = modifier_fitness(tmp.chromosomes)
        individus.append(tmp)
    return individus

def distance(V1, V2):
    d = (((V2[0] - V1[0])**2) + ((V2[1] - V1[1])**2))**(1/2)
    #return np.round(d,2)
    return d

def modifier_fitness(chromosomes):
    fitness = 0
    for i in range(0,len(chromosomes) - 1):
        d1 = Villes[chromosomes[i]]
        d2 = Villes[chromosomes[i+1]]
        fitness += distance(d1, d2)
    fitness= np.round(fitness, 2)
    return fitness

def meilleur_fitness(population_liste):
    fitness_liste=[]
    for i in population_liste:
        fitness_liste.append(i.fitness)
    min_fitness = min(fitness_liste)
    return population_liste[fitness_liste.index(min_fitness)]

def tournoi_selection(population,n):
    random_selections = random.choices(population, k=n)
    meilleurfitness = meilleur_fitness(random_selections)
    return meilleurfitness

def Reproduction(population):
    parent1 = tournoi_selection(population, nbVilles).chromosomes
    parent2 = tournoi_selection(population, nbVilles).chromosomes
    while parent1 == parent2:
        parent2 = tournoi_selection(population, 6).chromosomes
    return croisement(parent1, parent2)

def croisement(parent1, parent2):
    taille = len(parent1)
    fils = [-1] *taille
    fils[0], fils[taille - 1] = 0, 0
    point = random.randrange(1, taille-1)
    for i in range(point, point +1):
        fils[i] = parent1[i]
    point += 1
    point2 = point
    while fils[point] in [-1, 0]:
        if fils[point] != 0:
            if parent2[point2] not in fils:
                fils[point] = parent2[point2]
                point += 1
                if point == taille:
                    point = 0
            else:
                point2 += 1
                if point2 == taille:
                    point2 = 0
        else:
            point += 1
            if point == taille:
                point = 0
    if random.randrange(0, 100) < TAUXMUTATION:            
        fils = mutation(fils)
    p = Population()
    p.chromosomes = fils
    p.fitness = modifier_fitness(fils)
    return p

def mutation(chromosome):
        i1 = random.randrange(1,len(chromosome)-1)
        i2 = random.randrange(1,len(chromosome)-1)
        while i1 == i2:
            i2 = random.randrange(1, len(chromosome) - 1) #On génère une autre valeur de position
        temp = chromosome[i1]
        chromosome[i1] = chromosome[i2]
        chromosome[i2] = temp
        return chromosome
    
def trace(generation, fitness_list, meilleur_individu, Villes, nbVilles):
    plt.text((generation / 2) - 0.5, fitness_list[0] + 10, "nombre de générations: "+str(generation)+"\n meilleur Distance (fitness) : "+str(meilleur_individu.fitness) +"km")
    plt.plot(range(0, generation), fitness_list, c="red")
    plt.xlabel("nombre de générations ")
    plt.ylabel("Distance(KM)")
    if nbVilles>5 :
        print("Vous avez généré plusieurs villes, peut être que la figure est non lisible \n")
        choix = input("Voulez vous enregistrer la figure ?(O/N): ")
        if choix == 'O' or choix == 'o':
            plt.savefig("exampleVilles.pdf")
            print("fichier PDF enregistré dans le meme chemin du projet.")
    plt.show()
```

Maintenant, nous allons exploiter les fonctions définies précedemment pour résoudre le problème du voyageur de commerce sous différentes tailles.On selectionne le parent en effectuant un croisement puis on ajoute un nouveau enfant dans la population pour toutes les générations.
En même temps, on retire tout élement de la population crée et possèdant une valeur de fitness non valide<span style = "background-color : RGB(213, 221, 232);">(On garde que les valeurs minimale).</span>
Cette condition est vérifiée en comparant la valeur du fitness avec un <code>Seuil de Fitness</code> défini par défaut : 100 <br>

 Ainsi qu'on trace l'évolution de la distance minimale trouvée (fitness) en fonction de nombre de générations effectuées par l'algorithme.

```python
def Voyageur_de_commerce(nbVilles, n_gen):
    fitness_list = []
    population = creer_population(nbVilles)
    cmp_gen = 0
    for gen in range(n_gen):
        cmp_gen += 1
        for i in range(int(nbVilles/2)):
            population.append(Reproduction(population))
        for individu in population:
            if individu.fitness > 800:
                population.remove(individu)
        meilleur_individu = meilleur_fitness(population)
        print("=====================GENERATION========================", end = '\n')
        print("Géneration N° : ",cmp_gen, "\nFitness Minimale : ", meilleur_individu.fitness, "\nTaille de la Population : ", len(population),"\n")

        fitness_list.append(meilleur_individu.fitness)
    print("Liste des Fitness trouvée: " ,fitness_list)
    trace(cmp_gen,fitness_list, meilleur_individu, Villes,nbVilles)
    
n_gen = int(input("Entrez le nombre de génerations : "))
Voyageur_de_commerce(nbVilles, n_gen)
```

Finalement, on peut voir d'après le graphe que la distance optimale trouvé à la fin de chaque génération s'améliore et que la solution ait convergée vers le minimum trouvé pour ce nombre de génération, probable que l'augmentation de nombre des générations améliorera plus la solution si la solution la plus optimale n'est pas encore trouvée.
